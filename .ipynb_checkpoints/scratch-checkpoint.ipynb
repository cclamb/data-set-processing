```python
import h5py
import utilities
import sys
import ipdb

def viz(data):
    for row in data:
        for value in row:
            output = ' ' if value == 1 else '*'
            sys.stdout.write(output)
        sys.stdout.write('\n')
        
def row(data):
    min = 128
    max = 0
    cidx = 0
    for ridx in range(len(data)):
        minflag = False
        maxflag = False
        for cidx in range(len(data[0])):
            if data[ridx, cidx] != 1:
                if ridx < min and not minflag:
                    min = ridx
                    minflag = True
                if ridx > max and not maxflag:
                    max = ridx
                    maxflag = True
    return min, max

def col(data):
    min = 128
    max = 0
    cidx = 0
    for cidx in range(len(data[0])):
        minflag = False
        maxflag = False
        for ridx in range(len(data)):
            if data[ridx, cidx] != 1:
                if ridx < min and not minflag:
                    min = cidx
                    minflag = True
                if ridx > max and not maxflag:
                    max = cidx
                    maxflag = True
    return min, max
            
            

archive = h5py.File('archives/mnist-sd19.hdf5')

data = archive['train']['A']

print 'starting exam...'

rbounds = []
cbounds = []
for img in data:
    mat = img.reshape((128, 128))
    row_bounds = row(mat)
    col_bounds = col(mat)
    rbounds.append(row_bounds)
    cbounds.append(col_bounds)
    
print 'bounds generated...'

rminidx, cminidx = 0, 0
rmin, rmax = 128, 0
cmin, cmax = 128, 0
idx = 0
for bound in rbounds:
    if bound[0] < rmin:
        rminidx = idx
        rmin = bound[0]
    if bound[1] > rmax:
        rmax = bound[1]
    idx += 1
idx = 0
for bound in cbounds:
    if bound[0] < cmin:
        cminidx = idx
        cmin = bound[0]
    if bound[1] > cmax:
        cmax = bound[1]
    idx += 1
    
print 'bounds: %d %d %d %d, idx: %d %d .' % (rmin, rmax, cmin, cmax, rminidx, cminidx)
print 'col min ranges: %d %d' % (cbounds[cminidx][0], cbounds[cminidx][1])
print 'row min ranges: %d %d' % (rbounds[rminidx][0], rbounds[rminidx][1])

# ['train']['A'][2949] has row min, ['train']['A'][3745] has row max

archive.close()

# data = archive['train']['A'][2949]
# mat = data.reshape((128, 128))

# viz(mat)

```
